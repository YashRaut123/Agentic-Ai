### Streaming and Batch

## Streaming
Most models can stream thier output content while it is being generated. By displaying output progressively ,streaming significantly improves user experience,particularly for longer responses. Calling stream() returns an iterator that yield output chunks as they are produced. You can use a loop to process each chunk in real life. 

```python

import os
from langchain_google_genai import ChatGoogleGenerativeAI
model=ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
response=model.invoke("Write me a 200 words paragraph on Artificial intelligence")
response
```

```python
for chunk in model.stream("Write me a 200 words paragraph on Artificial intelligence"):
    print(chunk.text)
```

```python
for chunk in model.stream("Why go parrots have colorful feathers"):
    print(chunk.text, end="|" ,flush=True)
```

### Batch

Batching is a collection of independent requests to a model can significantly improve performance and reduce costs, as the processing can be done in parellel:

```python
responses = model.batch([
    "Who is Virat Kohli",
    "Tell me something about Lionel Messi",
    "Which club did Ronaldo played for in La Liga"
])
for response in responses:
    print(response)
```

```python
responses = model.batch([
    "Who is Virat Kohli",
    "Tell me something about Lionel Messi",
    "Which club did Ronaldo played for in La Liga"
],
config={
    'max_concurrency':5 #Limit to parellel calls
}
)
```
